```python
#4 Follow the instructions in the Exercise for creating a loyalty flag for existing customers using the transform() and loc() functions.

# Create a 'max_order' column to use for the 'loyalty_flag'

ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
# Checking to verify the column was created properly

ords_prods_merge.head(15)
```

```python
# Creating the loyalty_flag column criteria. 

ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
ords_prods_merge['loyalty_flag'].value_counts(dropna = False)
```

```python
# Checking the 3 columns used in the function

ords_prods_merge[['user_id', 'loyalty_flag', 'order_number']].head(25)
```

```python
#5 Use the loyalty flag you created and check the basic statistics of the product prices for each loyalty category (Loyal Customer, Regular Customer, and New Customer). 

ords_prods_merge.groupby('loyalty_flag').agg({'prices': ['mean', 'min', 'max']})
```

```python
# New customers appear to buy the highest priced items, followed by regular customers, and then the loyal customers buy the lowest priced items overall. This can be observed by looking at the mean of each category. 
```

```python
#6  Create a spending flag for each user based on the average price across all their orders using the following criteria:
# If the mean of the prices of products purchased by a user is lower than 10, then flag them as a “Low spender.”
# If the mean of the prices of products purchased by a user is higher than or equal to 10, then flag them as a “High spender.”

ords_prods_merge['average_prices'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
# Check to verify column was created properly

ords_prods_merge.head()
```

```python
# Creating the spending_flag column criteria. 

ords_prods_merge.loc[ords_prods_merge['average_prices']>= 10, 'spender_flag'] = 'High_spender'
ords_prods_merge.loc[ords_prods_merge['average_prices']<10, 'spender_flag'] = 'Low spender'
```

```python
ords_prods_merge['spender_flag'].value_counts(dropna = False)
```

```python
# Checking the columns used in the function

ords_prods_merge[['user_id', 'spender_flag', 'average_prices', 'prices']].head(25)
```

```python
#7 Create an order frequency flag that marks the regularity of a user’s ordering behavior according to the median in the “days_since_prior_order” column. The criteria for the flag should be as follows:
# If the median of “days_since_prior_order” is higher than 20, then the customer should be labeled a “Non-frequent customer.”
# If the median is higher than 10 and lower than or equal to 20, then the customer should be labeled a “Regular customer.”
# If the median is lower than or equal to 10, then the customer should be labeled a “Frequent customer.”

ords_prods_merge['median_frequency'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
# Check to verify column was created properly

ords_prods_merge.head()
```

```python
# Create the order_frequency column criteria

ords_prods_merge.loc[ords_prods_merge['median_frequency'] > 20, 'frequency_flag'] = 'Non-frequent customer'
ords_prods_merge.loc[(ords_prods_merge['median_frequency'] <= 20) & (ords_prods_merge['median_frequency'] > 10), 'frequency_flag'] = 'Regular customer'
ords_prods_merge.loc[ords_prods_merge['median_frequency'] <= 10, 'frequency_flag'] = 'Frequent customer'
```

```python
ords_prods_merge['frequency_flag'].value_counts(dropna = False)
```

```python
# Checking the columns used in the function

ords_prods_merge[['user_id', 'frequency_flag', 'median_frequency', 'days_since_prior_order']].head(25)
```
